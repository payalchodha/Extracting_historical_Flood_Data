```python
#function to web scrape Dainik Tribune
def scrape_dainik(soup_dnk):
    division = soup_dnk.find('div',{'class':'font_styl'})
    diction = {}
    diction['heading'] = division.find_all('p')
    diction['date']    = soup_dnk.find_all('span',{'class':'clock'})
    diction['source']  = "Dainik Tribune"
    return diction
```

```python
heading= []
for url in url_dainik_df['url']:
    res = requests.get(url)
    soup =BeautifulSoup(res.content,'lxml')
    heading.append(scrape_dainik(soup))
```

```python
#scraping function for NDTV
def ndtv_scraping(soup):
    diction = {}
    diction['heading'] =soup.find('meta',{'itemprop':"description"}).attrs['content']
    diction['date']    =soup.find('meta',{'itemprop':"uploadDate"}).attrs['content']
    diction['source']  = "NDTV"
   
    return diction
```

```python
for url in url_df_ndtv['url']:
  
    res = requests.get(url)
   
    soup =BeautifulSoup(res.content,'lxml')
    heading.append(ndtv_scraping(soup))
```
