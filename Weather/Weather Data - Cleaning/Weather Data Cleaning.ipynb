# Weather Data Cleaning

```python
import pandas as pd
import numpy as np
```

```python
#Importing our locations names, lat long files
latlong = pd.read_csv('./latlongpunjab.csv')
latlong['Latitude decimal'] = round(latlong['Latitude decimal'],3)
latlong['Longitude decimal'] = round(latlong['Longitude decimal'],3)
latlong['latlong'] = latlong['Latitude decimal'].map(str) + ','+ latlong['Longitude decimal'].map(str)
latlong
```

```python
latlong['Longitude decimal'][9] = 75.4432
```

```python
latlong
```

```python
#First combine both csvs together
weather = pd.read_csv("./raw_weatherdata/weather20082019.csv")
weather.shape
```

```python
weather.head()
```

```python
#remove duplicate lat/long
weather = weather.drop(["lat", "long"], axis=1)
```

```python
weather.isnull().sum()
```

```python
weather['Lon'].unique()
```

```python
#Need to drop Jalandhar because the longitude was wrong
weather = weather.drop(weather[weather.Lon < 50].index)
```

```python
weather.shape
```

```python
# Add City & clean up
longloc = latlong[['Longitude decimal','Location']]
```

```python
longloc
```

```python
weather = pd.merge(weather, longloc, left_on = 'Lon', right_on = "Longitude decimal", how = "left")
```

```python
weather.head()
```

```python
# For some reason, the name "Mansa" didn't work, it comes through as Null - will address later in the code
weather[weather["Location"].isnull()]['Lat'].value_counts()
```

```python
weather = weather.drop(["Longitude decimal"], axis=1)
```

```python
weather.head()
```

```python
weather = weather.rename(columns={'Past 1hour Precipitation(in)':'Past 1hour Precipitation(cm)'})
```

```python
#Unravel the min/max/sum/avg into separate columns
weather.head()
```

```python
weather['Past 1hour Precipitation(cm)'].apply
```

```python
#Inspired by: https://stackoverflow.com/questions/39899005/how-to-flatten-a-pandas-dataframe-with-some-columns-as-json
import ast
from pandas.io.json import json_normalize

def only_dict(d):
    
    return ast.literal_eval(d)


A = json_normalize(weather['Past 1hour Precipitation(cm)'].apply(only_dict).tolist()).add_prefix('PrecipPastHour(cm)_')
B = json_normalize(weather['Runoff(mm)'].apply(only_dict).tolist()).add_prefix('Runoff(mm)_')
C = json_normalize(weather['Soil Moisture% (0-10)'].apply(only_dict).tolist()).add_prefix('Soil Moisture%(0-10)_')
D = json_normalize(weather['Soil Moisture% (10-40)'].apply(only_dict).tolist()).add_prefix('Soil Moisture% (10-40)_')
E = json_normalize(weather['Soil Moisture% (40-100)'].apply(only_dict).tolist()).add_prefix('Soil Moisture% (40-100)_')
```

```python
A
```

```python
weather = weather[['Location', 'Lat','Lon','Year','Month','Day']].join([A, B, C, D, E])
```

```python
weather.Location.unique()
```

```python
weather['Location'].replace(to_replace = np.NaN, value='Mansa', inplace = True)
```

```python
weather.Location.unique()
```

```python
weather.shape
```

```python
weather.to_csv(path_or_buf = './cleaned_weatherdata/weather_CLEAN.csv', index=False)
```
